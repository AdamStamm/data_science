#Import de biblioteca

```
from pysus.online_data import SIH
```

#Acessando dados

```
dados = SIH.download('PR', 2020, '2')
dados.head()
```

É necessário informar no método download o estado, o ano (em inteiro) e o mês (em string) conforme acima.

```
dados.head()
```

```
#Definindo uma função para download de bases genéricas.
def baixando_dados_sih(estado, ano, mes):
  dados = SIH.download(estado, ano, mes)
  return dados
```

```
teste = baixando_dados_sih('PR', 2020, '2')
```

Percebi que existe um fator dificultante em utilizar a função pois o código fonte do pysus realizar um comando .upper() em uma lista de strings (no caso a de estados) e dessa forma o python retorna um erro. Vou estudar outras formas de automatizar essa extração posteriormente.

```
fev_2020 = baixando_dados_sih('PR', 2020, '2')
mar_2020 = baixando_dados_sih('PR', 2020, '3')
abr_2020 = baixando_dados_sih('PR', 2020, '4')
```

```
print(f'o número de linhas de fev_2020 eh: {fev_2020.shape[0]}')
print(f'o número de linhas de mar_2020 eh: {mar_2020.shape[0]}')
print(f'o número de linhas de abr_2020 eh: {abr_2020.shape[0]}')
```
